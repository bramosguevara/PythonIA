Así mismo, se desea correlacionar diferentes atributos externos al rendimiento escolar propiamente de los estudiantes, con el fin de conocer cómo esto implica en su rendimiento final del año, donde se tienen tres periodos académicos.

#### Los atributos para el dataset son:
- school - student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
- sex - student's sex (binary: 'F' - female or 'M' - male)
- age - student's age (numeric: from 15 to 22)
- address - student's home address type (binary: 'U' - urban or 'R' - rural)
- famsize - family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
- Pstatus - parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
- Medu - mother's education (numeric: 0 - none,  1 - primary education (4th grade), 2 â€“ 5th to 9th grade, 3 â€“ secondary education or 4 â€“ higher education)
- Fedu - father's education (numeric: 0 - none,  1 - primary education (4th grade), 2 â€“ 5th to 9th grade, 3 â€“ secondary education or 4 â€“ higher education)
- Mjob - mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- Fjob - father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' or 'other')
- reason - reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
- guardian - student's guardian (nominal: 'mother', 'father' or 'other')
- traveltime - home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
- studytime - weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
- failures - number of past class failures (numeric: n if 1<=n<3, else 4)
- schoolsup - extra educational support (binary: yes or no)
- famsup - family educational support (binary: yes or no)
- paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
- activities - extra-curricular activities (binary: yes or no)
- nursery - attended nursery school (binary: yes or no)
- higher - wants to take higher education (binary: yes or no)
- internet - Internet access at home (binary: yes or no)
- romantic - with a romantic relationship (binary: yes or no)
- famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
- freetime - free time after school (numeric: from 1 - very low to 5 - very high)
- goout - going out with friends (numeric: from 1 - very low to 5 - very high)
- Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
- Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
- health - current health status (numeric: from 1 - very bad to 5 - very good)
- absences - number of school absences (numeric: from 0 to 93)

#### Estos son los periodos: 
- G1 - first period grade (numeric: from 0 to 20)
- G2 - second period grade (numeric: from 0 to 20)
- G3 - final grade (numeric: from 0 to 20, output target)

De acuerdo a esto, podemos organizar los atributos con relación a cada categoría, para tener una mayor claridad de cuáles variables son las más interesantes a la hora del análisis exploratorio: 
- **Factores academicos:**

- **Factores personales y familiares:**
  - sex: Genero del estudiante

- **Factores sociales y estilo de vida:**
  - goout: Frecuencia en la que sale con amigos
  - Dalc y Walc: Consumo de alcohol entre semana y fines de semana
  - absences: Numero de ausencias a clases
  - freetime: Tiempo libre despues de escuela
  - famrel: Calidad de la relacion familiar

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga del archivo DataSet
DataFrame = pd.read_csv('student-mat.csv', sep=';')
print(DataFrame.shape)
DataFrame.head()
```

```python
#Visualizar las variables para conocer si existen valores nulos
print(DataFrame.info())
```

```python
#Conocer las estadisticas descriptivas del DataSet
DataFrame.describe()
```

```python
#Explorar si existen valores nulos en cada variable
print(DataFrame.isnull().sum())
```

### Analisis exploratorio de los datos de acuerdo a las variables más importantes

```python
#La idea es segmentar las variables de acuerdo con las más importantes referentes en cómo implican los factores
#En la nota final de matemáticas de acuerdo a la escuela.
DataFrame = DataFrame[['age', 'sex', 'studytime', 'failures', 'G1', 'G2', 'G3', 'famsup', 'Medu', 'Fedu', 'goout', 'Dalc', 'Walc', 'absences', 'school', 'freetime', 'famrel']]
```

```python
# Matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(DataFrame.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación entre Variables")
plt.show()
```

```python
# Rendimiento academico por género
print(DataFrame.groupby('sex')['G3'].mean())

plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['sex'], y=DataFrame['G3'])
plt.title("Rendimiento Académico por Género")
plt.xlabel("Género")
plt.ylabel("Nota Final (G3)")
plt.show()
```

**Se puede observar que en este caso las mujeres son las que tienen peor rendimiento en la asignatura de Matemáticas en relación con los hombres. Esto da una visibilidad mayor donde se observa que las mujeres tienen menos habilidad con las matemáticas de acuerdo a estas dos escuelas.**

```python
# Relacion entre el consumo de alcohol en el fin de semana y el rendimiento academico
print(DataFrame.groupby('Walc')['G3'].mean())
plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['Walc'], y=DataFrame['G3'])
plt.title("Consumo de Alcohol(Fin de Semana) y Nota Final")
plt.xlabel("Consumo de Alcohol en el fin de semana")
plt.ylabel("Nota Final (G3)")
plt.show()
# Relacion entre el consumo de alcohol en dias de la semana y el rendimiento academico
print(DataFrame.groupby('Dalc')['G3'].mean())
plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['Dalc'], y=DataFrame['G3'])
plt.title("Consumo de Alcohol(Dias Semana) y Nota Final")
plt.xlabel("Consumo de Alcohol en los dias de la semana")
plt.ylabel("Nota Final (G3)")
plt.show()
```

**Se puede observar que el consumo de alcohol puede interferir directamente en la nota final causando un impacto medio en el puntaje final, esto quiere decir que en el caso de cuando los estudiantes consumen alcohol tanto en el fin de semana como en los días de la semana se tiene un rango de nota final entre 9 y 10 queriendo decir que esto impacta negativamente, ya que está en el rango de reprobar la materia.**

```python
# Factores Externos
# Calcular la correlación de factores sociales con la nota final
social_factors = ['famrel', 'freetime', 'goout', 'failures', 'absences', 'Walc', 'Dalc']
correlations = DataFrame[social_factors + ['G3']].corr()['G3'].drop('G3')

# Mostrar el factor con mayor impacto en el rendimiento académico
most_impactful_factor = correlations.abs().idxmax()
print(f"El factor externo que más impacta en el rendimiento académico es: {most_impactful_factor}")

# Visualización de correlaciones
plt.figure(figsize=(6,4))
sns.barplot(x=correlations.index, y=correlations.values)
plt.title("Correlación de Factores Sociales con el Rendimiento Académico")
plt.xlabel("Factor externo")
plt.ylabel("Correlación con G3")
plt.show()
#Veces reprobadas
print(DataFrame.groupby('failures')['G3'].mean())
plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['failures'], y=DataFrame['G3'])
plt.title("Impacto de Factores Sociales en el Rendimiento Académico")
plt.xlabel("Numero de veces reprobadas")
plt.ylabel("Nota Final (G3)")
plt.show()
#Numero de ausencias
print(DataFrame.groupby('absences')['G3'].mean())
plt.figure(figsize=(8,4))
sns.barplot(x=DataFrame.groupby('absences')['G3'].mean().index, y=DataFrame.groupby('absences')['G3'].mean().values)
plt.title("Relación entre Número de Ausencias y Nota Final")
plt.xlabel("Número de Ausencias")
plt.ylabel("Promedio de Nota Final (G3)")
plt.show()
```

**En este apartado se puede observar que el factor que más impacta en la nota final del estudiante es la cantidad de veces reprobadas de la asignatura, ya que se relaciona directamente con la nota final de la asignatura. Donde se obtiene que cuando el estudiante ha visto la materia 3 o más veces, su nota final tiende a ser más baja, queriendo decir que el entendimiento propiamente en la asignatura se dificulta. <br>
Así mismo, el número de ausencias se relaciona directamente con la nota final, ya que para algunos estudiantes esto puede significar mayor o menor rendimiento en la asignatura, esto no es dependiente, ya que hay estudiantes que aun teniendo muchas ausencias su nota final es mejor con relación a otros.**


```python
# Análisis del nivel educativo de los padres y la nota final
plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['Medu'], y=DataFrame['G3'])
plt.title("Nivel Educativo de la madre y Nota Final")
plt.xlabel("Nivel Educativo de la Madre")
plt.ylabel("Nota Final (G3)")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['Fedu'], y=DataFrame['G3'])
plt.title("Nivel Educativo del padre y Nota Final")
plt.xlabel("Nivel Educativo del Padre")
plt.ylabel("Nota Final (G3)")
plt.show()
```

**La influencia del nivel educativo de la madre y del padre no es directamente superficial dentro del rendimiento del estudiante en la asignatura, pero si se puede notar una ligera atenuación por el nivel educativo de la madre donde se llegan a ver valores más altos en cuestión de la nota final con relación al nivel educativo del padre. Algunos estudiantes directamente no van relacionados al nivel educativo de los padres, ya que su rendimiento no se ve afectado por esta dependencia.**

```python
# Análisis de edades y rendimiento
high_performance = DataFrame[DataFrame['G3'] >= DataFrame['G3'].mean()]
age_counts = high_performance['age'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=age_counts.index, y=age_counts.values)
plt.title("Distribución de Edad de Estudiantes con Alto Rendimiento")
plt.xlabel("Edad")
plt.ylabel("Cantidad de Estudiantes")
plt.show()
```

```python
# Identificar la escuela con mejor rendimiento entre estudiantes con G3 alto
best_schools = high_performance['school'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=best_schools.index, y=best_schools.values)
plt.title("Escuelas con Mayor Cantidad de Estudiantes con Buen Rendimiento")
plt.xlabel("Escuela")
plt.ylabel("Cantidad de Estudiantes")
plt.show()
```

**Se puede observar que, en promedio, la mayoría de estudiantes con un mayor rendimiento en la asignatura se encuentran en una edad de 16 años. Esto implica que los estudiantes de esta edad tienden a tener un mejor entendimiento en la asignatura. <br>
Así mismo, la mayoría de estudiantes que tienen un mejor promedio en la asignatura corresponden directamente a la escuela Gabriel Pereira (GP), por lo que se deduce que en esta escuela existe una mejor calidad de enseñanza en el área de matemáticas.** 

```python
# Análisis del tiempo de estudio y veces reprobadas
plt.figure(figsize=(6,4))
sns.boxplot(x=DataFrame['studytime'], y=DataFrame['G3'])
plt.title("Tiempo de Estudio y Nota Final")
plt.xlabel("Tiempo de Estudio")
plt.ylabel("Nota Final (G3)")
plt.show()

failures = DataFrame['failures'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=failures.index, y=failures.values, hue=failures.index, legend=False)
plt.title("Distribución de Fracasos Académicos")
plt.xlabel("Número de veces reprobadas")
plt.ylabel("Cantidad de Estudiantes")
plt.show()
print(DataFrame['failures'].value_counts())
```

**Se observa que en promedio el tiempo de estudio sí impacta directamente en el rendimiento de la asignatura, ya que entre más el estudiante estudie sobre la materia, así mismo será su resultado final. Pero tampoco se debe suponer y dar por dado que hay estudiantes que no requieren de una gran cantidad de horas dedicadas para tener mejores notas en una asignatura, como se observa en el gráfico anterior referente a la relación del tiempo de estudio con la nota final es importante clarificar que hay estudiantes que dedican menos de dos horas (1) y tienen una nota final significativa. Sin embargo, por el promedio se obtiene que entre más horas se dedique al estudio de la asignatura, se tendrán mejores notas finales. <br>
Por otro lado, se deduce que la cantidad de estudiantes mayoritariamente no han tenido que repetir la asignatura y esto tiene un impacto positivo en la nota final.**

```python
#Identificar la escuela con mejor rendimiento promedio en la asignatura de Matemáticas
school_performance = DataFrame.groupby('school')['G3'].mean().sort_values(ascending=False)
print("\nEscuela con mejor rendimiento promedio en Matemáticas:")
print(school_performance)

plt.figure(figsize=(6,4))
sns.barplot(x=school_performance.index, y=school_performance.values)
plt.title("Promedio de Notas Finales por Escuela")
plt.xlabel("Escuela")
plt.ylabel("Promedio Nota Final (G3)")
plt.show()
```

**Se observa que directamente la escuela Gabriel Pereira(GP) es mejor en la asignatura de Matemáticas en comparación con Mousinho da Silveira(MS), esto puede significar muchas cosas, donde en términos generales tanto los estudiantes como las escuelas propiamente son mucho más dedicados a estudiar y a enfatizar los conocimientos propiamente en esta materia.**

```python
#Grafico comparativo entre atributos mas importantes
plt.figure(figsize=(10,6))
for school in DataFrame['school'].unique():
    school_data = DataFrame[DataFrame['school'] == school].groupby('absences')['G3'].mean()
    plt.plot(school_data.index, school_data.values, marker='o', label=f'Escuela {school}')
plt.title("Relación entre Nota Final y Ausencias por Escuela")
plt.xlabel("Número de Ausencias")
plt.ylabel("Promedio Nota Final (G3)")
plt.legend(title="Escuela")
plt.show()
```

## Conclusiones
- Según el análisis, la nota final (G3) es ligeramente menor en los estudiantes de género femenino en comparación con los estudiantes de género masculino. Esto se observa en la media de las notas finales de cada grupo y en el gráfico generado.
- Se identificó una correlación negativa entre el consumo de alcohol y la nota final, donde tanto para el consumo de alcohol en fines de semana como en los días de la semana tienden a estar asociados con un menor rendimiento académico. Donde claramente se observa que los estudiantes que consumen más alcohol suelen obtener calificaciones más bajas en la asignatura.
- En los factores sociales y/o externos analizados, la cantidad de veces que un estudiante ha reprobado la asignatura influye negativamente en el rendimiento académico. Otros factores como el tiempo libre y las salidas con amigos también muestran cierta influencia negativa en las calificaciones finales.
- Se observa que el número de ausencias a las clases de la asignatura infiere directamente en la nota final de los estudiantes, lo cual, dependiendo del número acumulado de inasistencias, tiende a disminuir la calificación final.
- Por otro lado, se observa una tendencia positiva en las notas finales de acuerdo al nivel educativo de los padres, en el cual entre mejor sea el nivel educativo de los padres, así mismo tiende a mejorar el rendimiento de los estudiantes. Además, analizando las escuelas, se identifica que la escuela con mejor rendimiento en la asignatura de Matemáticas es la que tiene mayor promedio de nota final (G3) que tienen sus estudiantes, que en este caso es Gabriel Pereira (GP).
